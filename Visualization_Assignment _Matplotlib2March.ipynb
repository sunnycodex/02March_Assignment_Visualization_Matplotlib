```python
#2.
#Scatter plot is the plot where the intersect pooint does not connect by lines.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
plt.scatter(x,y)
plt.show()
```

```python
#3.
"""The subplot() function in Matplotlib's Pyplot module is used to create a grid of subplots within a single figure. 
It allows you to arrange multiple plots (axes) in a grid-like structure, making it easier to visualize and compare different plots side by side. 
The subplot() function is particularly useful when you want to create a figure with multiple subplots, such as multiple charts or graphs, in a single image."""
```

```python
#3.
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.subplot(2,2,2)
plt.plot(x2,y2)


plt.subplot(2,2,3)
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.show()
```

```python
#4.
#A bar plot is a type of data visualization that uses rectangular bars to represent data.
# uses of bar plot :
"""Comparison of Categorical Data: Bar plots are most commonly used to compare the values of different categories or groups. 
For example, you can use a bar plot to compare sales performance for different products, 
compare the scores of students in various subjects, or compare the popularity of different programming languages

Visualizing Frequency Distributions: Bar plots can be used to display the frequency or count of items within categories. 
This is especially useful for visualizing categorical data in the form of a histogram, 
where each bar represents a category and its height represents the frequency of occurrences.

Comparison of Multiple Groups: Bar plots can also be used to compare multiple groups or conditions side by side. For example, 
you can compare the average scores of students from different schools or the sales performance of different regions
"""
```

```python
#4.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.show()
```

```python
#4. For bar horizontal 
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company, profit)
plt.show()
```

```python
#5.
"""A box plot, also known as a box-and-whisker plot, is a statistical data visualization technique used to summarize and display the distribution of a dataset. 
It provides a graphical representation of the data's median, quartiles, and potential outliers. 
Box plots are particularly useful when comparing multiple datasets or visualizing the spread of a single dataset."""


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data= [box1,box2]

plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot for box1 and box2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data for box1 and box2
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a single list for plotting
data = [box1, box2]

# Create the box plot
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot for box1 and box2')
plt.show()
```
